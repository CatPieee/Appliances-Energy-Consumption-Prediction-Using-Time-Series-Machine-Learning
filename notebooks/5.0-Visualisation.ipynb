Visualisation

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6) # default figure size
plt.rcParams['font.size'] = 12           # default font size
plt.rcParams['font.family'] = 'Times New Roman'
```

Import prediction result data

```python
ML_results = pd.read_csv("../results/ML_predictions.csv")
prophet_results = pd.read_csv("../results/prophet_predictions.csv")
arima_results = pd.read_csv("../results/ARIMA_predictions.csv")
rnn_results = pd.read_csv("../results/rnn_predictions.csv")
lstm_results = pd.read_csv("../results/lstm_predictions.csv")
transformer_results = pd.read_csv("../results/transformer_predictions.csv")
```

Extract the minimum size of result data (Obtain the common test set part of all models)

```python
min_size = min(len(ML_results), len(prophet_results), len(arima_results), len(rnn_results), len(lstm_results), len(transformer_results))
print("Minimum size of result data: ", min_size)
```

Extract minimum size from result data

```python
ML_results_min = ML_results.tail(min_size)
prophet_results_min = prophet_results.tail(min_size)
arima_results_min = arima_results.tail(min_size)
rnn_results_min = rnn_results.tail(min_size)
lstm_results_min = lstm_results.tail(min_size)
transformer_results_min = transformer_results.tail(min_size)
```

Visualisation

Import ablation metrics

```python
ML_metrics          = pd.read_csv("../results/ablation_study/ML_metrics.csv")
ML_metrics_ablation = pd.read_csv("../results/ablation_study/ML_metrics_ablation.csv")
```

Ablation visualisation

```python
metric_names = ['MSE', 'RMSE', 'MAE', 'R2']
model_names = ML_metrics.iloc[:, 0]

fig, axes = plt.subplots(2, 2, figsize=(15, 12)) # Adjust figure size for better readability

bar_width = 0.35
index = np.arange(len(model_names))

for i, metric in enumerate(metric_names):
    row = i // 2
    col = i % 2

    # Get values for both scenarios
    metric_values_full = ML_metrics[metric].values
    metric_values_ablation = ML_metrics_ablation[metric].values

    # Determine bar colors
    if metric in ['MSE', 'RMSE', 'MAE']:
        color_full = 'firebrick'
        color_ablation = 'lightcoral'
    else: # R2
        color_full = 'forestgreen'
        color_ablation = 'lightgreen'

    # Plot bars
    bars_full = axes[row, col].bar(index - bar_width/2, metric_values_full, bar_width, label='Full Features', color=color_full, alpha=0.8, edgecolor='grey')
    bars_ablation = axes[row, col].bar(index + bar_width/2, metric_values_ablation, bar_width, label='Ablated Features', color=color_ablation, alpha=0.8, edgecolor='grey')

    axes[row, col].set_title(f'{metric} Comparison', fontsize=14, fontweight='bold', pad=15)
    axes[row, col].set_ylabel(metric, fontsize=12)
    axes[row, col].set_xticks(index)
    axes[row, col].set_xticklabels(model_names, rotation=35, ha='right') # Set x-axis labels to model names
    axes[row, col].grid(True, alpha=0.3, axis='y')
    axes[row, col].legend(fontsize=10)

    # Set y-axis limits
    all_values = np.concatenate([metric_values_full, metric_values_ablation])
    if metric in ['MSE', 'RMSE', 'MAE']:
        axes[row, col].set_ylim(0, max(all_values) * 1.15) # Increased limit for better text placement
    elif metric == 'R2':
        axes[row, col].set_ylim(min(0, min(all_values) * 1.1), 1) # Adjust min for negative R2

    # Add value labels on top of the bars
    for bar_idx, bar_full in enumerate(bars_full):
        height_full = bar_full.get_height()
        value_full = metric_values_full[bar_idx]
        # Adjust text position based on whether it's a positive or negative value for R2
        if metric == 'R2' and value_full < 0:
            axes[row, col].text(bar_full.get_x() + bar_full.get_width() / 2., height_full - (max(all_values) * 0.05),
                                 f'{value_full:.4f}', ha='center', va='top', fontsize=9, fontweight='bold', color='black')
        else:
            axes[row, col].text(bar_full.get_x() + bar_full.get_width() / 2., height_full + (max(all_values) * 0.01),
                                 f'{value_full:.4f}', ha='center', va='bottom', fontsize=9, fontweight='bold', color='black')

    for bar_idx, bar_ablation in enumerate(bars_ablation):
        height_ablation = bar_ablation.get_height()
        value_ablation = metric_values_ablation[bar_idx] # Corrected to use ablation values
        # Adjust text position based on whether it's a positive or negative value for R2
        if metric == 'R2' and value_ablation < 0:
            axes[row, col].text(bar_ablation.get_x() + bar_ablation.get_width() / 2., height_ablation - (max(all_values) * 0.05),
                                 f'{value_ablation:.4f}', ha='center', va='top', fontsize=9, fontweight='bold', color='black')
        else:
            axes[row, col].text(bar_ablation.get_x() + bar_ablation.get_width() / 2., height_ablation + (max(all_values) * 0.01),
                                 f'{value_ablation:.4f}', ha='center', va='bottom', fontsize=9, fontweight='bold', color='black')
                                 

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for suptitle
plt.suptitle('Machine Learning Models Performance Metrics Comparison with and without Feature Ablation', fontsize=16, fontweight='bold', y=0.98)
plt.savefig('../results/prediction_plots/ML_model_performance_ablation_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
